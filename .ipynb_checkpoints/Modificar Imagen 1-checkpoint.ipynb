<h1>Procesando Imagenes con Numpy</h1>

```python
import matplotlib as plt
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
import urllib.request
from PIL import Image

# para plotear con fondo oscuro
plt.style.use("dark_background")
```

<h4>Creamos un array con una imagen de internet</h4>

```python
url = 'https://www.goodnet.org/photos/620x0/31501_hd.jpg'
internet_image = Image.open(urllib.request.urlopen(url))
```

```python
#transformamos la imagen en un array
#preparamos 3 array por separadospara usarlos m√°s adelante en diferentes casos

image1 = np.array(internet_image)
image2 = np.array(internet_image)
image3 = np.array(internet_image)
image4 = np.array(internet_image)
image5 = np.array(internet_image)
```

<h2>Empezamos</h2>

```python
#veamos como luce la imagen
plt.figure(figsize=(10,20))
plt.imshow(image1)
```

```python
#la forma de la imagen (resoluci√≥n)
print(image1.shape)
649 * 1240 * 3
```

```python
#vemos el tama√±o de la imagen total
print(image1.size)
```

```python
#vemos los valores maximos y minimos de cada pixel
#en imagen RGB (Red Green Blue)
print(np.amin(image1), np.amax(image1))
```

```python
print("Fila Completa")
print(image1[324]) #imprimo una fila
print("_________________________________")
print("")
print(image1[324].shape)
```

```python
print("Columna Completa")
print(image1[:,620]) #imprimo columna
print("_____________________")
```

```python
plt.figure(figsize=(5,10))
plt.imshow(image1[150:400,400:650])
```

```python
#los 3 valores de un pixel
print("cantidad de valores en la imagen")
print(image1.shape)
print("3 colores de un pixel")
print(image1[324,620])
print("____________________")
```

```python
##Un solo valor de un pixel
print("3 colores de un pixel")
print(image1[324,620,0])
print("____________________")
```

<h1>Modificando</h1>

```python
#seleccionamos un rango de las filas y le ponemos el color rojo (indice 0 del RGB)
#al maximo valor que es 255
image2[0:100, : , 0] = 204
image2[0:100, : , 1] = 0
image2[0:100, : , 2] = 204
plt.figure(figsize=(10,20))
plt.imshow(image2)
```

```python
# vamos a crear filtros. seleccionando pixeles de terminados

filtro = image3 > 229
print(filtro.shape)
print(image3.shape)
```

```python
image3[filtro] -= 25
plt.figure(figsize=(10,20))
plt.imshow(image3)
```

```python

filtro2 = image4 > 50
image4[filtro2] -= 100
plt.figure(figsize=(10,20))
plt.imshow(image4)
```

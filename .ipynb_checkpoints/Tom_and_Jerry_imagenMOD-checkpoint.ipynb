```python
import matplotlib as plt
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt
import urllib.request
from PIL import Image

# para plotear con fondo oscuro
plt.style.use("dark_background")
```

```python
tom_and_jerry_image = Image.open("Tom_and_Jerry.jpg")
```

```python
image1 = np.array(tom_and_jerry_image)
```

```python
plt.figure(figsize=(10,20))
plt.imshow(image1)
```

```python
print(image1.shape)
```

```python
fila_x, colu_y, layersTotal = image1.shape
print(fila_x, colu_y, layersTotal)
```

```python
x, y = np.ogrid[:fila_x,:colu_y]
print(y)
print(x)
```

```python
center_x1, center_y1 = int(fila_x*.66), int(colu_y*.33)
center_x2, center_y2 = int(fila_x*.33), int(colu_y*.66)
print(center_x1, center_y1)
print(center_x2, center_y2)
```

```python
dot_uno = (x - center_x1)**2 + (y - center_y1)**2
dot_dos = (x - center_x2)**2 + (y - center_y2)**2
print(dot_uno)
print("_______________________________________")
print(dot_dos)
```

```python
#sacar el radio del circulo
circle_1 = (100)**2
circle_2 = (150)**2
print(circle_1)
print(circle_2)
```

```python
circle_Mask1 = (dot_uno < circle_1)
circle_Mask2 = (dot_dos < circle_2)
```

```python
image1[circle_Mask1 + circle_Mask2]=0
plt.figure(figsize=(10,20))
plt.imshow(image1)
```

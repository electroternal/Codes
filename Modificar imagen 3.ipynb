```python
url = 'https://www.goodnet.org/photos/620x0/31501_hd.jpg'
internet_image = Image.open(urllib.request.urlopen(url))
```

```python
#transformamos la imagen en un array
#preparamos 3 array por separadospara usarlos m√°s adelante en diferentes casos
image1 = np.array(internet_image)
image2 = np.array(internet_image)
image3 = np.array(internet_image)
```

```python
plt.figure(figsize=(10,20))
plt.imshow(image1)
```

```python
#creamos una mascara que tomar√° valor true, donde existan valores mallores a 115 en verdes
green_mask = image1[:,:,1] > 115
print(green_mask)

#operaremos sobre los pixeles a donde hay verdes intensos
image1[green_mask,0] =0
image1[green_mask,1] =0
image1[green_mask,2] =0

print(image1[green_mask])
plt.figure(figsize=(10,20))
plt.imshow(image1)
```
